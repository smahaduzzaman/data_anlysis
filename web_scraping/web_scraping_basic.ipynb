```python
class_name = "wikitable sortable"
table_list = soup.find_all("table", class_=class_name)
# print(len(table_list), *[i.text for i in table_list])
table_data = table_list[0]
table_data
```

```python
table_headers = table_data.find_all("th")
for header in table_headers:
  print(header.text)
```

```python
table_headers = table_data.find_all("th")
headers = []
for header in table_headers:
  headers.append(header.text.strip())
```

```python
table_row_data = []
for row in table_data.find_all("tr"):
  table_row_data.append(row)
  print(table_row_data)
```
